```python
correlation_data = numeric_data.corr()
```

```python
plt.figure(figsize = (8, 6))
sns.heatmap(correlation_data, annot = True, cmap = 'coolwarm', fmt = '.2f')
plt.title('Correlation_data Plot')
plt.show()
```

```python
# removing column names excluding 'Weight'
features = numeric_data.columns[numeric_data.columns != 'weight']
 
# Plot scatter plots for each feature against 'Weight'
plt.figure(figsize = (12, 8))
for feature in features:
    plt.scatter(numeric_data[feature], numeric_data['weight'], label = feature, alpha = 0.7)
```

```python
features = numeric_data.columns[numeric_data.columns != 'weight']
plt.figure(figsize = (12, 8))
for feature in features:
    plt.scatter(numeric_data[feature], numeric_data['weight'], label = feature, alpha = 0.7)
plt.title('Scatter Plots of Weight Against Other Variable')
plt.ylabel('Weight')
plt.xlabel('Other Variable')
plt.legend()
plt.show()
```

```python
features
```

```python
#Scatter plot of 'Weight' against 'Age'
plt.figure(figsize = (15, 10))
sns.scatterplot(y = 'weight', x = 'age', data = numeric_data)
plt.title('Scatter Plot of Weight Against Age')
plt.ylabel('Weight')
plt.xlabel('Age')
plt.show()
```

```python
#Looping through columns and plot histograms.
for column in numeric_data.columns:
    plt.figure(figsize = (8, 6))
    plt.hist(numeric_data[column], bins = 25, edgecolor = 'black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
features = numeric_data.columns[numeric_data.columns !='Weight']
#Looping through columns and plot scatter plots against 'Weight'
for feature in features:
    plt.figure(figsize = (8, 6))
    plt.scatter(numeric_data[feature], numeric_data['weight'])
    plt.title(f'Scatter Plot of {feature} Againt Weight')
    plt.xlabel(feature)
    plt.ylabel('weight')
    plt.show()
```

```python
# select all columns except 'Weight' and 'Age'
features = numeric_data.columns.difference(['weight', 'age'])
#Looping through columns and create bar plots against 'Weight', color-coded by 'Age'
for feature in features:
    plt.figure(figsize = (10, 6))
    sns.barplot(x = feature, y = 'weight', hue = 'age', data = numeric_data)
    plt.title(f'Bar Plot of {feature} Againt Weight (Color by Age)')
    plt.xlabel(feature)
    plt.ylabel('weight')
    plt.legend(title = 'age')
    plt.show()
```
